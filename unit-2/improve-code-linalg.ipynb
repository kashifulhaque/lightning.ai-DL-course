### **For-loops and dot products**

```python
b = 0.
x = [1.2, 2.2]
w = [3.3, 4.3]

output = b
for xj, wj in zip(x, w):
    output += xj * wj

output
```

```python
b = torch.tensor([0.])
x = torch.tensor([1.2, 2.2])
w = torch.tensor([3.3, 4.3])

x.dot(w) + b
```

### **Benchmark**

```python
b = 0.
x = [random.random() for _ in range(10_000)]
w = [random.random() for _ in range(10_000)]
```

```python
t_b = torch.tensor(b)
t_x = torch.tensor(x)
t_w = torch.tensor(w)
```

```python
def for_loop(x, w, b):
    output = b
    
    for xj, wj in zip(x, w):
        output += xj * wj
    
    return output
```

```python
def pytorch_dot(x, w, b):
    return x.dot(w) + b
```

```python
%timeit for_loop(x, w, b)
```

```python
%timeit pytorch_dot(t_x, t_w, t_b)
```

### **Dealing with multiple training examples via matrix multiplication**

```python
b = 0.
X = [
    [1.2, 2.2],
    [4.4, 5.5]
]
w = [3.3, 4.3]

outputs = []

for x in X:
    output = b

    for xj, wj in zip(x, w):
        output += xj * wj
    
    outputs.append(output)

outputs
```

```python
b = torch.tensor([0.])
X = torch.tensor([
    [1.2, 2.2],
    [4.4, 5.5]
])
w = torch.tensor([3.3, 4.4])

X.matmul(w) + b
```

### **Benchmark**

```python
b = 0.
X = [
    [random.random() for _ in range(10_000)]
    for i in range(1_000)
]
w = [random.random() for _ in range(10_000)]
```

```python
def for_loop(X, w, b):
    outputs = []

    for x in X:
        output = b

        for xj, wj in zip(x, w):
            output += xj * wj

        outputs.append(output)

    return outputs
```

```python
%timeit for_loop(X, w, b)
```

```python
t_b = torch.tensor(b)
t_X = torch.tensor(X)
t_w = torch.tensor(w)
```

```python
def pytorch_implementation(X, w, b):
    return X.matmul(w) + b
```

```python
%timeit pytorch_implementation(t_X, t_w, t_b)
```

## **Broadcasting: computations and unequal tensor shapes**

```python
a = torch.tensor([1.1, 2.1, 3.1, 4.1])
b = torch.tensor([5.4, 5.5, 5.6, 5.7])

a + b
```

```python
A = torch.tensor([[1.1, 2.1, 3.1, 4.1],
                  [1.2, 2.2, 3.2, 4.2]])
b = torch.tensor([5.4, 5.5, 5.6, 5.7])

A + b
```
